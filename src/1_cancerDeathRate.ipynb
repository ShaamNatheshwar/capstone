# Colon and Rectum Cancer Death by Year

```python
plt.figure(figsize=(16,8))
df.groupby('Year').sum()['Colon and rectum cancer'].plot(linewidth=4)
plt.tight_layout()
plt.title('Colon and rectum cancer death by Year', size = 18)
plt.ylabel("Colon and rectum cancer")
plt.xlabel("Year")
```

```python
Colon_Top = df.groupby('Entity').sum()['Colon and rectum cancer'].nlargest(10)
Colon_Top = pd.DataFrame(Colon_Top)
```

# Countries with highest number of Colon and Rectum Cancer Death

```python
plt.figure(figsize=(16,8))
sns.barplot(x=Colon_Top['Colon and rectum cancer'],y=Colon_Top.index,data=Colon_Top,palette='Spectral')
plt.title('Top countries with highest number of Colon and rectum cancer death',fontsize=18, color = 'black')
plt.ylabel('Country',fontsize=15)
plt.xlabel('Colon and rectum cancer',fontsize=15)
```

# Colon and Rectum Cancer death by Year in USA

```python
america = df.query(' Entity == "United States" ')[['Entity', 'Year','Colon and rectum cancer']]
fig = plt.figure(figsize=(27,18))

sub1 = fig.add_subplot(2,2,1)
sub1.set_xlabel('Year')
sub1.set_title('Colon and rectum cancer death by Year in USA')
america.groupby('Year').sum()['Colon and rectum cancer'].plot(linewidth=4,color='blue')
plt.legend(bbox_to_anchor=(1, 1), loc=1)
```

# Colon and Rectum Cancer death by Year in India

```python
India = df.query(' Entity == "India" ')[['Entity', 'Year','Colon and rectum cancer']]
fig = plt.figure(figsize=(27,18))

sub1 = fig.add_subplot(2,2,1)
sub1.set_xlabel('Year')
sub1.set_title('Colon and rectum cancer death by Year in India')
India.groupby('Year').sum()['Colon and rectum cancer'].plot(linewidth=4,color='blue')
plt.legend(bbox_to_anchor=(1, 1), loc=1)
```

# Recent Trends in Colon and Rectum Cancer Death Rate (as of 2019)

```python
cancer2019 = df.query(' Year == 2019 ')
Coloncancer2019 = cancer2019.sort_values(by = 'Colon and rectum cancer', ascending=False).head(10)
plt.figure(figsize=(16,8))
sns.barplot(x=Coloncancer2019['Colon and rectum cancer'],y=Coloncancer2019['Entity'],data=Coloncancer2019,palette='RdGy')
plt.title('Top countries with highest number of Colon and rectum cancer death in 2019',fontsize=18, color = 'black')
plt.ylabel('Country',fontsize=15)
plt.xlabel('Colon cancer',fontsize=15)
```
